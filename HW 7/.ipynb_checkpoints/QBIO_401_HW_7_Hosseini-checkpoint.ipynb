```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("corrafterrain.bracken",sep="\t",skiprows=0,header=1)
df
```

```python
num_mapped_reads = sum(df["new_est_reads"])
num_mapped_reads
```

4,884,486 reads were mapped to a species from our metagenomics sample.

```python
num_species = df.shape[0]
num_species
```

There were 4,539 species identified in this sample.

Next we are going to randomly sub-sample reads from this dataset to see where we are in the
refraction curve. A rarefaction curve plots the number of reads on the x-axis and the number of
species on the y-axis. As you can imagine, if we have a very small number reads then additional
reads will yield many new species, while if we have a very large number of reads then
additional reads will yield few new species (instead these additional reads will usually belong to
species we have already found).

**I want you to write a function to estimate how many different species we would have found if
we had sequenced only a fraction f of the reads we actually sequenced. I want you to report the
estimated number of different species for the following values of the fraction f: 0.0001, 0.001,
0.01, 0.1, 0.5, and 1.0 (1.0 means we sequenced all the reads we actually sampled). In addition,
I want you to make a plot with the f values on the x-axis and the estimated number of different
species on the y-axis. What are the values of f where we would have found about the same
number of different species we found with the full dataset? What are the values of f where we
would have found many fewer species than we found with the full dataset?**

```python
f_values_vector = [0.0001, 0.001, 0.01, 0.1, 0.5, 1.0]
pj_values_vector = df['new_est_reads'] / num_mapped_reads
pj_values_vector
```

```python
rng = np.random.default_rng()
poisson_simul = []
for f in f_values_vector:
    poisson_simul.append(rng.poisson(pj_values_vector*num_mapped_reads*f, num_species))
poisson_simul
```

```python
num_diff_species = []
for simul in poisson_simul:
    num_diff_species.append(sum(x > 0 for x in simul))
num_diff_species
```

```python
for i in range(len(f_values_vector)):
    print(f"For f = {f_values_vector[i]}, we would have found {num_diff_species[i]} different species in our metagenomics sample.")
```

```python
plt.plot(f_values_vector, num_diff_species)
plt.xlabel("f-value")
plt.ylabel("estimated number of different species found")
plt.title("Relationship between fraction of reads actually sequenced (f) and estimated # of diff. species found")
```

```python
plt.plot(f_values_vector, num_diff_species)
plt.xlabel("f-value")
plt.ylabel("estimated number of different species found")
plt.title("Relationship between fraction of reads actually sequenced (f) and estimated # of diff. species found")
plt.xscale('log')
```

```python
plt.bar([str(x) for x in f_values_vector], num_diff_species)
plt.xlabel("f-value")
plt.ylabel("estimated number of different species found")
plt.title("Relationship between fraction of reads actually sequenced (f) and estimated # of diff. species found")
```

The f-values for which we would have found about the same number of different species in our metagenomics sample are 0.1, 0.5, and 1. The f-values for which we would have observed significanlty less different species in our sample are 0.01, 0.001, and 0.0001. For 0.01, we observe about 75% as much diversity in species found. For 0.001, we would observe only about 33% as much diversity in species found. For 0.0001, wqe would observe a mere 5% as much diversity in the species found. 

**To condense the above analysis into a function which estimates how many different species we would have found if we had sequenced only a fraction f of the reads we actually sequenced:**

```python
def estimate_num_diff_species(bracken_df, f_values_vector: list) -> list:
    num_mapped_reads = sum(bracken_df["new_est_reads"])
    num_species = bracken_df.shape[0]
    pj_values_vector = bracken_df['new_est_reads'] / num_mapped_reads
    rng = np.random.default_rng()
    poisson_simul = []
    for f in f_values_vector:
        poisson_simul.append(rng.poisson(pj_values_vector*num_mapped_reads*f, num_species))
    num_diff_species = []
    for simul in poisson_simul:
        num_diff_species.append(sum(x > 0 for x in simul))
    for i in range(len(f_values_vector)):
        print(f"For f = {f_values_vector[i]}, we would have found {num_diff_species[i]} different species in our metagenomics sample.")
```

```python
estimate_num_diff_species(df, f_values_vector)
```
